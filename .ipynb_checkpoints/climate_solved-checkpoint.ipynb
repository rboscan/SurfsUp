# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session, join
from sqlalchemy import create_engine, func, inspect
from datetime import datetime, timedelta
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Climate Analysis

```python
## Design a query to retrieve the last 12 months of precipitation data and plot the results
# Calculate the date 1 year ago from the last data point in the database
last_date = session.query(Measurement.date).order_by(Measurement.date.desc()).first()
first_date = datetime(2016,8,23)
# Perform a query to retrieve the data and precipitation scores
year_prcp = session.query(Measurement.prcp,Measurement.date) .\
    filter(Measurement.date >= first_date).all()
# Save the query results as a Pandas DataFrame and set the index to the date column
year_prcp_df = pd.DataFrame(year_prcp)
# Sort the dataframe by date
year_prcp_df['date']=year_prcp_df['date'].to_numpy(dtype='datetime64[ns]')
year_prcp_df.sort_values(by=['date'],ascending=True)
year_prcp_df=year_prcp_df.set_index('date')
year_prcp_df.columns=['Precipitation']
# Use Pandas Plotting with Matplotlib to plot the data
fig, ax = plt.subplots(figsize=(10,8))
year_prcp_df.plot(ax = ax, x_compat = True)

ax.set_xlabel('Date')
ax.set_ylabel('Precipitation (in.)')
ax.set_title('Honolulu, Hawaii Precipitation 2016 to 2017')
ax.set_xticklabels([])

plt.tight_layout()
plt.savefig('Images/Honolulu-2016-17.png')
```

```python
# Design a query to show how many stations are available in this dataset?
station_count = session.query(Station.id).distinct().count()
station_count
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.
inspector = inspect(engine)
inspector.get_table_names()
station_columns = inspector.get_columns('Station')
measurement_columns = inspector.get_columns('Measurement')

stations = session.query(Station.station, func.count(Measurement.id)).\
    join(Station, Measurement.station == Station.station).group_by(Station.station).\
    order_by(func.count(Measurement.id).desc()).all()

for result in stations:
    print(f"Station: {result[0]}  Count: {result[1]}")
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature most active station?
active_station = 'USC00519281'
station_temps = session.query(func.max(Measurement.tobs), func.min(Measurement.tobs),func.avg(Measurement.tobs)).\
                filter(Measurement.station == active_station)
print(f"Highest Temp: {station_temps[0][0]} \xb0F")
print(f"Lowest Temp: {station_temps[0][1]} \xb0F")
print(f"Average Temp: {round(station_temps[0][2], 1)} \xb0F")
```

```python
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
active_station_temps = session.query(Measurement.date, Measurement.tobs).filter(Measurement.station==active_station).\
    filter(Measurement.date >= first_date).all()

station_temps_df = pd.DataFrame(active_station_temps, columns = ['Date','Temperature'])
station_temps_df = station_temps_df.set_index('Date', drop=True)
station_temps_df.head()
```

```python
fix, ax = plt.subplots()
station_temps_df.plot.hist(bins=12,ax = ax,figsize=(6,5))
ax.set_ylabel('Frequency')
ax.set_xlabel('Temperature (\xb0F)')
plt.tight_layout()
plt.savefig('Images/histogram.png')
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Use your previous function `calc_temps` to calculate the tmin, tavg, and tmax 
# for your trip using the previous year's data for those same dates.
trip = calc_temps('2016-05-05','2016-05-17')

print(f'Highest temperature during trip {trip[0][2]} \xb0F')
print(f'Lowest temperature during trip {trip[0][0]} \xb0F')
print(f'Average temperature during trip {round(trip[0][1],1)} \xb0F')
```

```python
# Plot the results from your previous query as a bar chart. 
# Use "Trip Avg Temp" as your Title
# Use the average temperature for the y value
# Use the peak-to-peak (tmax-tmin) value as the y error bar (yerr)
peak_to_peak = trip[0][2] - trip[0][0]
average = trip[0][1]

fig, ax = plt.subplots(figsize=(3,6))

ax.bar(1, average, yerr = peak_to_peak/2, width = .5)
ax.set_title('Trip Avg Temp')
ax.set_ylabel('Temperature \xb0F')

plt.tight_layout()
plt.savefig('Images/Trip-Avg-Temp.png')
```

```python
# Calculate the total amount of rainfall per weather station for your trip dates using the previous year's matching dates.
# Sort this in descending order by precipitation amount and list the station, name, latitude, longitude, and elevation

rainfall = session.query(Station.station, Station.name, Station.latitude, Station.longitude,
                               Station.elevation, func.avg(Measurement.prcp)).\
    filter(Measurement.station == Station.station).\
    filter(func.strftime("%Y-%m-%d", Measurement.date) >= start_date).group_by(Station.station).\
    order_by(func.avg(Measurement.prcp).desc()).all()

rainfall_df = pd.DataFrame(rainfall, columns=['Station','Name','Latitude','Longitude','Elevation','Precipitation Avg. (in)'])

rainfall_df
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")
```

```python
# calculate the daily normals for your trip
# push each tuple of calculations into a list called `normals`
# Set the start and end date of the trip
start_date = datetime(2016,5,5)
end_date = datetime(2016,5,17)
# Use the start and end date to create a range of dates
date_range = pd.date_range(start=start_date, end=end_date,periods=12)
# Stip off the year and save a list of %m-%d strings
no_year=[]
for date in date_range:
    no_year.append(date.strftime('%m-%d'))
no_year
# Loop through the list of %m-%d strings and calculate the normals for each date
trip_max=[]
trip_min=[]
trip_avg=[]
for date in no_year:
    trip_max.append(daily_normals(date)[0][2])
    trip_min.append(daily_normals(date)[0][0])
    trip_avg.append(daily_normals(date)[0][1])
```

```python
# Load the previous query results into a Pandas DataFrame and add the `trip_dates` range as the `date` index
trip_temps_df = pd.DataFrame({'Max Temp':trip_max, 
                              'Min Temp':trip_min, 
                              'Avg Temp':trip_avg,
                              'Date':no_year})
trip_temps_df = trip_temps_df.set_index('Date')
trip_temps_df
```

```python
# Plot the daily normals as an area plot with `stacked=False`
ax = trip_temps_df.plot.area(stacked=False)
ax.set_xticklabels(labels=no_year,rotation=45)
plt.tight_layout()
plt.plot()
plt.savefig('Images/daily-normals.png')
```
